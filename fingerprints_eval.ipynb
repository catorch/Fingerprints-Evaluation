# Fingerprints

1. Extended-Connectivity Fingerprints (ECFP): A type of circular fingerprint, widely used in cheminformatics for similarity searches and structure-activity modeling.
    * 1024 bits
    * 2048 bits
    * \>2048 bits

2. **MACCS Keys**: A set of 166 structural fragment keys used to represent molecular structure.

3. **Daylight Fingerprints**: Linear fingerprints based on the Daylight chemical information systems approach, involving atom pairs and paths.

4. **PubChem Fingerprints**: A set of predefined patterns and substructure keys used to represent chemical compounds, as used by the PubChem database.

5. **Atom Pair Fingerprints**: Represents a molecule by pairwise topological distances between atoms.

6. **Topological Torsion Fingerprints**: Encodes topological information based on torsion angles around bonds.

7. **Morgan Fingerprints**: Another term for ECFP, often used in the RDKit cheminformatics toolkit.

8. **Lingo Fingerprints**: Based on substrings of SMILES notations, representing small fragments of the molecule.

9. **Pharmacophore Fingerprints**: Encodes the presence of pharmacophore features like hydrogen bond donors/acceptors, aromatic rings, etc.

10. **GpiDAPH3 Fingerprints**: A three-dimensional fingerprint encoding distances between pharmacophore features.

11. **Shape Signatures**: Represents the three-dimensional shape of molecules, useful in similarity searches.

12. **Electrotopological State (E-State) Indices**: Captures electronic and topological information around each atom.

13. **2D Autocorrelation Descriptors**: Encodes the distribution of certain properties (e.g., charge, mass) along the 2D structure.

14. **3D Property Fingerprints**: Involves interatomic distances, molecular surfaces, and electrostatic fields.

15. **Neural Fingerprints**: Designed by convolutional neural networks to optimize fingerprints for specific property sets.

16. **MinHash Fingerprints**: An approach for efficiently comparing high-dimensional ECFP-style fingerprints using hashing techniques.

tldr; There is not necessarily a "best" fingerprint method because there are many, many ways to generate fingerprints. ECFP4 or ECFP6 with >2048 bits is usually decent.

Generally speaking, a fingerprint is some large vector, typically binary, used for similarity-based screening and more recently machine learning. The elements of the fingerprint are some type of chemical descriptor.

Some recent reviews:

"Machine learning in chemoinformatics and drug discovery" Drug Discovery Today 2018, v. 23, August pp. 1538-1546
"Similarity Searching Using 2D Structural Fingerprints" Methods Mol Biol. 2011 v. 672 pp. 133-58.
IMHO, there are a few general categories:

Structural pattern fingerprints (e.g. MACCS) - sets of functional group patterns (e.g. alkyl chains, aromatic rings, ketones, ..) one bit or count per functional group in the fingerprint.
Daylight-style linear fingerprints - take the atom, and it's linear neighbors
Circular-style fingerprints (e.g. ECFP) - generating a radius or diameter around the neighbors of an atom. Like the linear fingerprints, these can essentially generate an infinite number of patterns, so it's hashed to a fixed bit vector
3D property fingerprints - interatomic distances, surfaces, electrostatic fields, etc.
There are countless new fingerprint schemes produced each year, e.g.

"Neural Fingerprints" https://github.com/HIPS/neural-fingerprint - convolutional neural networks to design the best fingerprints for a particular property set
MinHash fingerprints https://github.com/reymond-group/mhfp - a better hashing scheme for the high-dimensional ECFP-style circular fingerprint
Open Babel can compute a few pattern-style fingerprints (FP3, FP4, and MACCS), the Daylight-style linear FP2 fingerprint, and a variety of ECFP circular fingerprint styles, e.g. ECFP4.

RDKit can generate a whole pile of fingerprints too: http://rdkit.org/docs/GettingStartedInPython.html#fingerprinting-and-molecular-similarity

You might say, but is there a good place to start? The Morgan circular / ECFP fingerprints are really good, especially with a fairly large bit vector (e.g. 2048 bits per molecule) to hash.
